```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
data = pd.read_excel('youtube.xlsx')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.shape
```

```python
data.Likes.fillna(value = data.Likes.mean(), inplace = True)
data.Likes.fillna(value = data.Comments.mean(), inplace = True)
data.Views.fillna(value = data.Views.mean(), inplace = True)
```

### Which keyword has  high number of videos?

```python
data.Keyword.value_counts()
```

```python
plt.figure(figsize = (15, 7))
data.groupby('Keyword')['Video ID'].count().plot(kind = 'bar', color = 'orange')
plt.show()
```

### Which keyword has high likes?

```python
data.groupby('Keyword')['Likes'].sum().sort_values(ascending = False)
```

```python
plt.figure(figsize = (15, 6))
data.groupby('Keyword')['Likes'].sum().plot(kind = 'bar')
plt.show()
```

### Which keyword has high comments?

```python
data.groupby('Keyword')['Comments'].sum().sort_values(ascending = False)
```

```python
plt.figure(figsize = (15, 6))
data.groupby('Keyword')['Comments'].sum().plot(kind = 'bar', color = 'green')
plt.show()
```

### Which keyword has high comments?

```python
data.groupby('Keyword')['Views'].sum().sort_values(ascending = False)
```

```python
plt.figure(figsize = (15, 6))
data.groupby('Keyword')['Views'].sum().plot(kind = 'bar', color = 'black')
plt.show()
```
